###**2- Criação e configuração do Namenode**

Um Namenode é um componente central em um cluster Hadoop responsável por gerenciar o sistema de arquivos distribuído chamado **Hadoop Distributed File System (HDFS)**. Suas funções incluem manter o controle do sistema de arquivos, controlar o acesso, gerenciar blocos de dados e monitorar a saúde dos servidores de armazenamento de dados chamados DataNodes.

###**2.4- Criando a Imagem via docker**

Uma imagem é um pacote completo e autossuficiente que contém todos os componentes necessários para executar um aplicativo. Ela serve como um modelo para criar contêineres, permitindo que os aplicativos sejam executados de forma consistente e isolada em ambientes Docker.

**A**- Certifiquei-me de que os arquivos necessários estavam na pasta de trabalho, conforme mencionado. Em seguida, abri o terminal e naveguei até a pasta do namenode onde eu tinha meu arquivo Dockerfile. Depois, executei o seguinte comando para criar a imagem:

*** *sudo docker build . -t namenode:ds*a***

Isso criou uma imagem chamada **"namenode"** com a tag **"dsa"** a partir das instruções definidas no Dockerfile. Agora, eu estou pronto para usar essa imagem para criar e executar um contêiner NameNode do Hadoop.

###**2.7- Acessar o container Namenode para configuração interna**

Configurações esperadas:

* Formatar o **Namenode** **(somente na primeira execução)**;

Essas configurações são necessárias, pois após subir o container do **Namenode** para garantir que o ambiente esteja configurado corretamente e pronto para ser usado no cluster Hadoop essas configurações  foram executadas. Essas etapas são parte do processo de configuração inicial do **cluster Hadoop HDFS** e da preparação do ambiente para execução.





**A**- Configurando Internamente o Container Namenode\
|**Service ssh restart**|

A Instrução **"sudo service ssh restart"** é usada para reiniciar o serviço SSH dentro do container. É necessária porque, durante a construção do container e da imagem, foram feitas configurações relacionadas ao **SSH**, como a geração de **chaves SSH** e a configuração do acesso sem senha.\
Portanto, executar **"sudo service ssh restart"** é importante para garantir que o **SSH** esteja configurado e funcionando corretamente dentro do container, o que pode ser necessário para acessar e gerenciar o container remotamente.

Para entrarmos dentro do container, podemos usar o comando.\
***sudo docker container exec -it nome do container/bin/sh***, Ou optar por usar Container console do portainer.io

Usando o comando acima, ao entrar no container digite **bash** e em seguida o comando\
***sudo service ssh restart***

ssh2.png






**B**- Configurando Internamente o Container Namenode\
**|Ajuste dos privilégios**|

Fiz os ajustes de privilégios para conceder ao usuário **"hduser"** o controle total sobre os diretórios relevantes dentro do container. Isso é fundamental para garantir que o ambiente **Hadoop HDFS** funcione corretamente, pois o **Hadoop HDFS** requer permissões específicas para acessar e gerenciar seus diretórios de trabalho, armazenamento e outros recursos. Dessa forma, ao ajustar os privilégios e a propriedade dos diretórios, assegurei que o **Hadoop** possa executar suas operações sem problemas dentro do contêiner.

executados os comandos abaixo.

***sudo chown -R hduser:hduser/home/hduser/jdk***\
***sudo chown -R hduser:hduser/home/hduser/hadoop***

Resultados:
Troca do usuario root, para Hduser em jdk e Hadoop

**C**- Configurando Internamente o Container Namenode\
**|Formatar o Container Namenode|**

Formatei o nameNode do **Hadoop HDFS**, especialmente durante a primeira execução, pois é um passo crítico na configuração inicial de um cluster Hadoop HDFS por várias razões:
* Inicialização Limpa;
* Criação de Metadados Iniciais;
* Prevenção de Conflitos de Metadados;
* Segurança e Isolamento

Comando utlizado para formatar o Namenode.\
***hdfs namenode -format***

**D**- Configurando Internamente o Container Namenode\
**|Iniciar o Namenode|**

Comando utiizado para iniciar o **Namenode**

***hdfs --daemon start namenode***

Se precisar parar o serviço:\
**hdfs --daemon stop namenode**

Obs: Após digitar o comando para iniciar o Namenode, **digitei jps**. Esse comando listou todos os processos Java em execução, incluindo o **NameNode**, e exibiu os **PIDs** correspondentes. O número exibido corresponde ao **PID** do processo do Namenode. Foi uma forma de verificar se o serviço do NameNode estava em execução, o que foi útil para o monitoramento e gerenciamento do **cluster Hadoop HDFS**.
